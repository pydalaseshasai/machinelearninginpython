```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(corrmat, vmax=.8, square=True)
```

```python
sns.catplot(x="Age", y="Sex", data=df, kind="boxen",height=4, aspect=2.7, hue = "Embarked")
```

```python

g = sns.FacetGrid(df, row="Embarked", hue="Embarked", aspect=5, height=3)
g.map(sns.kdeplot, "Age", shade=True, alpha=1, lw=3.5, bw=.2)
g.map(sns.kdeplot, "Age", color="w", lw=2, bw=.2)
g.map(plt.axhline, y=0, lw=2)
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, color=color, ha="left", va="center", transform=ax.transAxes)
g.map(label, "age")

g.fig.subplots_adjust(hspace=-.25)

g.set_titles("")
g.set(yticks=[])
g.despine(bottom=True, left=True)
```
